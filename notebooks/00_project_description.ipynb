# 00 – Student Success: Factors and Insights

---

## <u>**Objective**</u>

This project aims to explore the key factors that influence student academic performance using data-driven analysis.
The dataset contains information from <u>**6,590**</u> students, including study habits, attendance, parental involvement, motivation level, sleep hours, and access to resources.
By analyzing these factors, we seek to identify patterns and relationships that contribute to higher academic success.

---

## <u>**Research Questions**</u>

1. Does the number of <u>**study hours**</u> and <u>**sleep hours**</u> affect student performance?
2. Is there a correlation between <u>**parental involvement**</u> and <u>**student attendance**</u>?
3. Do <u>**extracurricular activities**</u> and <u>**motivation levels**</u> contribute to better academic results?
